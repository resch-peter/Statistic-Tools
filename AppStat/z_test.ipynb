```python
def chi2_test(hist1, hist2):
    """ Compute the chi2 test statistic between two histograms. """
    chi2 = 0
    for i in range(len(hist1)):
        if hist1[i] + hist2[i] > 0:
            chi2 += (hist1[i] - hist2[i])**2 / (hist1[i] + hist2[i])
    return chi2
```
