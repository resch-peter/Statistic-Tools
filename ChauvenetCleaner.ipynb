```python
import numpy as np
from scipy import special   
```

```python
for k in range(0,4):
    mask=mask_with[k]
    data=Ppos_with[k]
    #print(data)
    #print(mask)


    chauv_fact=1/2
    b=2
    continue_cleaning=True


    while continue_cleaning:
        d=data[mask]
        mean_pre=d.mean()
        std_pre=d.std()#ddof=len(data))
        continue_cleaning=False

        print(d)
        print(mean_pre, std_pre, len(d))

        if(b==0 or std_pre<1.0): 
            print("stop")
            break

        for i in range(len(data)):
            if mask[i] and special.erfc(abs(data[i]-mean_pre)/std_pre)*len(d)<chauv_fact:
                mask[i]=False
                continue_cleaning=True
                print("reject",data[i])

        #print("done with iteration", b, "; continue_cleaning:", continue_cleaning)
        #print()
        b=b-1

    mask_without[k]=mask
    print(f"Data after Chauvenet Cleaning Procedure - Mean: {mean_pre:.3f} +- {std_pre/np.sqrt(len(d)):.3f}     (Std = {std_pre:.3f})")
    print(f"  The number of measurements used before and after is: {len(data):d} and {len(d):d}")
    print()
    print()
```
