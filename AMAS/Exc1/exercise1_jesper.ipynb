```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from scipy import stats

sns.set(style="darkgrid", rc={"font.family": "serif"})
# sns.set_context("notebook", font_scale=1.5)

sns.set_palette("gnuplot")
np.random.seed(123)
```

```python
numbers1 = np.genfromtxt("FranksNumbers.txt", skip_header=2, max_rows=11)
numbers2 = np.genfromtxt("FranksNumbers.txt", skip_header=2 + 11 + 2, max_rows=11)
numbers3 = np.genfromtxt(
    "FranksNumbers.txt", skip_header=2 + 11 + 2 + 11 + 2, max_rows=11
)
numbers4 = np.genfromtxt(
    "FranksNumbers.txt", skip_header=2 + 11 + 2 + 11 + 2 + 11 + 2, max_rows=11
)
numbers5 = np.genfromtxt(
    "FranksNumbers.txt",
    skip_header=2 + 11 + 2 + 11 + 2 + 11 + 2 + 11 + 2,
    max_rows=99 - 55,
)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(numbers1[:, 0], numbers1[:, 1], label="Set 1", marker="o", linestyle="None")
plt.plot(numbers2[:, 0], numbers2[:, 1], label="Set 2", marker="o", linestyle="None")
plt.plot(numbers3[:, 0], numbers3[:, 1], label="Set 3", marker="o", linestyle="None")
plt.plot(numbers4[:, 0], numbers4[:, 1], label="Set 4", marker="o", linestyle="None")
# plt.plot(numbers5[:, 0], numbers5[:, 1], label='Set 5',marker='o', linestyle='None')
plt.legend()
plt.show()
```

```python
mean1 = np.mean(numbers1[:, 1])
meanx1 = np.mean(numbers1[:, 0])
varx1 = np.var(numbers1[:, 0])
var1 = np.var(numbers1[:, 1])
mean2 = np.mean(numbers2[:, 1])
meanx2 = np.mean(numbers2[:, 0])
varx2 = np.var(numbers2[:, 0])
var2 = np.var(numbers2[:, 1])
mean3 = np.mean(numbers3[:, 1])
meanx3 = np.mean(numbers3[:, 0])
varx3 = np.var(numbers3[:, 0])
var3 = np.var(numbers3[:, 1])
mean4 = np.mean(numbers4[:, 1])
meanx4 = np.mean(numbers4[:, 0])
varx4 = np.var(numbers4[:, 0])
var4 = np.var(numbers4[:, 1])
mean5 = np.mean(numbers5[:, 1])
meanx5 = np.mean(numbers5[:, 0])
varx5 = np.var(numbers5[:, 0])
var5 = np.var(numbers5[:, 1])
```

```python
fig, ax = plt.subplots(5, 1, figsize=(8, 20))
ax[0].plot(numbers1[:, 0], numbers1[:, 1], marker="o", linestyle="None")
ax[0].errorbar(
    meanx1,
    mean1,
    xerr=np.sqrt(varx1),
    yerr=np.sqrt(var1),
    fmt="o",
    label="Mean ± Std Dev = {:.2f} ± {:.2f}".format(mean1, np.sqrt(var1)),
)
ax[0].set_title("Set 1")
ax[0].legend()
ax[1].plot(numbers2[:, 0], numbers2[:, 1], marker="o", linestyle="None")
ax[1].errorbar(
    meanx2,
    mean2,
    xerr=np.sqrt(varx2),
    yerr=np.sqrt(var2),
    fmt="o",
    label=f"Mean ± Std Dev = {mean2:.2f} ± {np.sqrt(var2):.2f}",
)
ax[1].set_title("Set 2")
ax[1].legend()
ax[2].plot(numbers3[:, 0], numbers3[:, 1], marker="o", linestyle="None")
ax[2].errorbar(
    meanx3,
    mean3,
    xerr=np.sqrt(varx3),
    yerr=np.sqrt(var3),
    fmt="o",
    label="Mean ± Std Dev = {:.2f} ± {:.2f}".format(mean3, np.sqrt(var3)),
)
ax[2].set_title("Set 3")
ax[2].legend()
ax[3].plot(numbers4[:, 0], numbers4[:, 1], marker="o", linestyle="None")
ax[3].errorbar(
    meanx4,
    mean4,
    xerr=np.sqrt(varx4),
    yerr=np.sqrt(var4),
    fmt="o",
    label="Mean ± Std Dev = {:.2f} ± {:.2f}".format(mean4, np.sqrt(var4)),
)
ax[3].set_title("Set 4")
ax[3].legend()
ax[4].plot(numbers5[:, 0], numbers5[:, 1], marker="o", linestyle="None")
ax[4].errorbar(
    meanx5,
    mean5,
    xerr=np.sqrt(varx5),
    yerr=np.sqrt(var5),
    fmt="o",
    label="Mean ± Std Dev = {:.2f} ± {:.2f}".format(mean5, np.sqrt(var5)),
)
ax[4].set_title("Set 5")
ax[4].legend()
plt.tight_layout()
plt.show()
```

Part 2

```python
expected1 = numbers1[:, 0] * 0.48 + 3.02
expected2 = numbers2[:, 0] * 0.48 + 3.02
expected3 = numbers3[:, 0] * 0.48 + 3.02
expected4 = numbers4[:, 0] * 0.48 + 3.02
expected5 = numbers5[:, 0] * 0.48 + 3.02


def chi_squared(observed, expected, error):
    return np.sum((observed - expected) ** 2 / error**2)


chi2_1 = chi_squared(numbers1[:, 1], expected1, np.sqrt(numbers1[:, 1]))
chi2_2 = chi_squared(numbers2[:, 1], expected2, np.sqrt(numbers2[:, 1]))
chi2_3 = chi_squared(numbers3[:, 1], expected3, np.sqrt(numbers3[:, 1]))
chi2_4 = chi_squared(numbers4[:, 1], expected4, np.sqrt(numbers4[:, 1]))
chi2_5 = chi_squared(numbers5[:, 1], expected5, np.sqrt(numbers5[:, 1]))
print(f"Reduced Chi-squared Set 1: {chi2_1 / len(numbers1[:, 1]):.2f}")
print(f"Reduced Chi-squared Set 2: {chi2_2 / len(numbers2[:, 1]):.2f}")
print(f"Reduced Chi-squared Set 3: {chi2_3 / len(numbers3[:, 1]):.2f}")
print(f"Reduced Chi-squared Set 4: {chi2_4 / len(numbers4[:, 1]):.2f}")
print(f"Reduced Chi-squared Set 5: {chi2_5 / len(numbers5[:, 1]):.2f}")
```

Now using an error of +/-1.22

```python
chi2_1_e2 = chi_squared(numbers1[:, 1], expected1, 1.22)
chi2_2_e2 = chi_squared(numbers2[:, 1], expected2, 1.22)
chi2_3_e2 = chi_squared(numbers3[:, 1], expected3, 1.22)
chi2_4_e2 = chi_squared(numbers4[:, 1], expected4, 1.22)
chi2_5_e2 = chi_squared(numbers5[:, 1], expected5, 1.22)
print(f"Reduced Chi-squared Set 1 (error=1.22): {chi2_1_e2 / len(numbers1[:, 1]):.2f}")
print(f"Reduced Chi-squared Set 2 (error=1.22): {chi2_2_e2 / len(numbers2[:, 1]):.2f}")
print(f"Reduced Chi-squared Set 3 (error=1.22): {chi2_3_e2 / len(numbers3[:, 1]):.2f}")
print(f"Reduced Chi-squared Set 4 (error=1.22): {chi2_4_e2 / len(numbers4[:, 1]):.2f}")
print(f"Reduced Chi-squared Set 5 (error=1.22): {chi2_5_e2 / len(numbers5[:, 1]):.2f}")
```
