```python
import numpy as np
import seaborn as sns
from iminuit import Minuit

sns.set(style="darkgrid", rc={"font.family": "serif"})
sns.set_palette("gnuplot2")

np.random.seed(123)
```

```python
x_values = np.genfromtxt("MC_ex_2_lec_3.txt")


def norm_pdf(var, a, b):
    return 1 / (2 * (1 + b / 3)) * (1 + a * var + b * var**2)


def neg_ln_likelihood(samples, a, b):
    return -np.sum(np.log(norm_pdf(samples, a, b)))


m = Minuit(lambda a, b: neg_ln_likelihood(x_values, a, b), a=0, b=0)
m.migrad()
```

```python
# Trying unnormalized pdf to see that it does not work
def unnorm_pdf(var, a, b):
    return 1 + a * var + b * var**2


def neg_ln_likelihood_unnorm(samples, a, b):
    return -np.sum(np.log(unnorm_pdf(samples, a, b)))


X_wrong = []
while len(X_wrong) < 2000:
    x = np.random.uniform(-1, 1)
    u = np.random.uniform(0, 2)
    if u < unnorm_pdf(x, 0.5, 0.5):
        X_wrong.append(x)

X_wrong = np.array(X_wrong)
m_wrong = Minuit(lambda a, b: neg_ln_likelihood_unnorm(X_wrong, a, b), a=0, b=0)
m_wrong.migrad()
```
